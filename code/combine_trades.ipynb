```python
import pandas as pd
import glob
import os
```

```python
currencies = pd.read_csv('../temporary/currencies.csv')
```

```python
symbols = list(currencies['symbol'][currencies['altcoin'] == False]) #Remove other cryptocurrencies and extract currency code/symbol

#create list of files
files = []
for symbol in symbols:
    filename = 'trades_' + str(symbol) + '.csv'
    files.append(filename)
```

```python
#load and combine files in chunks
CHUNK_SIZE = 50000 #how many rows to load at once
output_file = '../temporary/trades_combined.csv'

#check if combined file is already there, if so delete

if os.path.exists(output_file):
    os.remove(output_file)
    print("The file has been deleted successfully.")
else:
    print("File does not exist, no action necessary.")

first_file = True
for file in files: # if it is not the first csv file then skip the header row (row 0) of that file
    
    if not first_file:
        skip_row = [0]
        head = True
    else:
        skip_row = []
        head = False
        
    path = '../temporary/trades_by_country/' + file
    print(file)
    if os.path.exists(path):
        try:
            chunk_container = pd.read_csv(path, chunksize = CHUNK_SIZE, skiprows = skip_row)
            for chunk in chunk_container:
                chunk.to_csv(output_file, mode = 'a', index = False)
        except pd.errors.EmptyDataError:
            print('This file: ' + file + ' has no data in it!')
    
    first_file = False
        
```
